# Impute missing value in Rating with Median

```python
# impute value with median
df['Rating'] = df['Rating'].fillna((df['Rating'].median()))
```

# Imputation in Current version, Android version, type (only one null value) with MODE

```python
# replace with mode

df['Type'] = df['Type'].fillna((df['Type'].mode()[0]))
df['Current Ver'] = df['Current Ver'].fillna((df['Current Ver'].mode()[0]))
df['Android Ver'] = df['Android Ver'].fillna((df['Android Ver'].mode()[0]))
```

```python
# replace with mode but with loop method
col = ['Type', 'Current Ver', 'Android Ver']
for i in col:
    df[i] = df[i].fillna((df[i].mode()[0]))
    
```
